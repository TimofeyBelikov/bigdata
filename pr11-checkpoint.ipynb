### Подключение библиотек

```python
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import tree
from sklearn import ensemble

warnings.filterwarnings('ignore')
```

### Подключение данных

```python
df = pd.read_csv('House_Rent_Dataset.csv')
data = df.drop(['Posted On', 'BHK', 'Floor', 'Area Type', 'Area Locality', 'City', 'Furnishing Status', 'Tenant Preferred', 'Bathroom', 'Point of Contact'], axis=1)
display(data.head())

scaler = MinMaxScaler()
scaler.fit(data)
data[['Rent', 'Size']] = scaler.transform(data)
plt.scatter(data['Size'], data['Rent'])
plt.show()
# Удалим выброс
data = data.drop(data[data.Rent > 0.8].index)
plt.scatter(data['Size'], data['Rent'])
plt.show()
```

### Дерево решений на подвыборках

```python
models = []
for i in range(10):
    train = data.loc[np.random.choice(data.index, int(len(data)/10), replace=False)]
    model_tree = tree.DecisionTreeRegressor(max_depth=8, random_state=1)
    model_tree.fit(np.array(train['Size']).reshape(-1,1), train['Rent'])
    models.append(model_tree)
models
```

```python
x = np.linspace(0.0, 1.0, len(data))
y_pred = []
for i in range(len(models)):
    y_pred.append(models[i].predict(x.reshape(-1,1)))
```

```python
plt.xlabel('Size')
plt.ylabel('Rent')
for i in range(len(models)):
    plt.scatter(x, y_pred[i], c= 'red', s = 2)
plt.scatter(data['Size'], data['Rent'], c='black', s=2)
plt.show()
```

```python
mean_pred = np.array(y_pred).mean(axis=0)
print(mean_pred)
plt.xlabel('Size')
plt.ylabel('Mean Rent')
plt.scatter(data['Size'], data['Rent'], c='black', s=2)
plt.scatter(x, mean_pred, c = 'green', s=2)
plt.show()
```

### На всём наборе данных

```python
model_tree = tree.DecisionTreeRegressor(max_depth=8, random_state=1)
one_model = model_tree.fit(np.array(data['Size']).reshape(-1,1), data['Rent'])
one_pred = one_model.predict(x.reshape(-1,1))

plt.xlabel('Size')
plt.ylabel('Rent')
plt.scatter(data['Size'], data['Rent'], c='black', s=2)
plt.scatter(x, one_pred, c = 'orange', s =2)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.ensemble import RandomForestRegressor
import catboost as cb
```

```python
x_train, x_test, y_train, y_test = train_test_split(data['Size'], data['Rent'], train_size = 0.8)
print(x_train.shape)
print(x_test.shape)

random_forest = RandomForestRegressor(max_depth=10, min_samples_split=10).fit(np.array(x_train).reshape(-1,1), np.array(y_train).reshape(-1,1))
y_pred_t = random_forest.predict(np.array(x_train).reshape(-1,1))
print('MSE для тренировочных данных', mean_squared_error(y_pred_t, y_train))
y_pred = random_forest.predict(np.array(x_test).reshape(-1,1))
print('MSE для тестовых данных', mean_squared_error(y_pred, y_test))
```

```python
model_catboost_reg = cb.CatBoostRegressor(iterations=3000, task_type='CPU', devices='0')
model_catboost_reg.fit(x_train, y_train)
```

```python
y_pred_t = model_catboost_reg.predict(np.array(x_train).reshape(-1,1))
print('MSE для тренировочных данных', mean_squared_error(y_pred_t, y_train))
y_pred = model_catboost_reg.predict(np.array(x_test).reshape(-1,1))
print('MSE для тестовых данных', mean_squared_error(y_pred, y_test))
```

```python
x = np.linspace(0.0, 1.0, len(data))
one_pred_catboost = model_catboost_reg.predict(x.reshape(-1,1))

plt.xlabel('Size')
plt.ylabel('Rent')
plt.scatter(data['Size'], data['Rent'], c='black', s=2)
plt.scatter(x, mean_pred, c = 'green', s=2)
plt.scatter(x, one_pred_catboost, c = 'red', s =2)
plt.show()
```
